## Load the tags deemed useful

```python
with open("tags.txt", "r", encoding="utf-8") as file:
    tags_set = {line.strip() for line in file}
```

```python
tags_set
```

```python
def has_matching_tag(tag_string):
    tags = {tag for tag in tag_string.split(",")}
    return tags.issubset(tags_set)
```

```python
filtered_df = df[df["RelatedTagsArray"].apply(has_matching_tag)]
```

```python
def majority_tags_in_set(tag_string, threshold=0.8):
    tags = {tag.strip() for tag in tag_string.split(",")}
    if not tags:
        return False
    matches = tags.intersection(tags_set)
    return len(matches) / len(tags) >= threshold

filtered_df = df[df["RelatedTagsArray"].apply(lambda x: majority_tags_in_set(x, 0.65))]
```

### Copy and save

```python
import os
import shutil

# Ensure output directory exists
os.makedirs("firstpass", exist_ok=True)

# Iterate over rows in the filtered DataFrame
for _, row in filtered_df.iterrows():
    domain_id = row["domain_id"]
    page_id = row["page_id"]
    page_version_id = row["page_version_id"]
    
    # Build the source and destination file paths
    filename = f"{domain_id}_{page_id}_{page_version_id}.html"
    src_path = os.path.join("html", filename)
    dst_path = os.path.join("firstpass", filename)
    
    # Copy the file if it exists
    if os.path.isfile(src_path):
        shutil.copy2(src_path, dst_path)
    else:
        print(f"Missing: {src_path}") 
```

```python
import os
import shutil
from bs4 import BeautifulSoup
from langdetect import detect, LangDetectException
from tqdm import tqdm

# Ensure output directory exists
os.makedirs("engpass", exist_ok=True)

# Iterate over the DataFrame with a progress bar
for _, row in tqdm(filtered_df.iterrows(), total=len(filtered_df), desc="Processing HTML files"):
    domain_id = row["domain_id"]
    page_id = row["page_id"]
    page_version_id = row["page_version_id"]
    
    filename = f"{domain_id}_{page_id}_{page_version_id}.html"
    src_path = os.path.join("firstpass", filename)
    dst_path = os.path.join("engpass", filename)

    if not os.path.isfile(src_path):
        continue

    try:
        with open(src_path, "r", encoding="utf-8", errors="ignore") as f:
            html = f.read()

        soup = BeautifulSoup(html, "html.parser")
        text = soup.get_text(separator=" ", strip=True)

        if text and detect(text) == "en":
            shutil.copy2(src_path, dst_path)

    except (LangDetectException, UnicodeDecodeError):
        continue 
```

```python
from bs4 import BeautifulSoup
import hashlib
```

```python
def extract_text_and_hash(path):
    try:
        with open(path, "r", encoding="utf-8", errors="ignore") as f:
            html = f.read()
        text = BeautifulSoup(html, "html.parser").get_text(" ", strip=True)
        return hashlib.sha256(text.encode("utf-8")).hexdigest()
    except Exception as e:
        return None 

df["hash"] = df["file"].apply(extract_text_and_hash)

unique_hashes = df["hash"].nunique()
print(f"Number of distinct hashes: {unique_hashes}")

# optionally see groups
hash_groups = df.groupby("hash")["file"].apply(list)
```

```python
import os
import glob
import pandas as pd

# Step 1: list all files in engpass/*.html
all_files = glob.glob("engpass/*.html")
all_files = [s.replace('\\', '/') for s in all_files]
# Step 2: load CSV
df = pd.read_csv("labeled_unbalanced_final.csv")

# normalize CSV file paths (strip spaces just in case)
csv_files = df["file"].astype(str).str.strip().tolist()

# Step 3: find files not in CSV
missing_files = sorted(set(all_files) - set(csv_files))

# Step 4: write to file_list.js
with open("file_list.js", "w", encoding="utf-8") as f:
    f.write("const filesToLabel = [\n")
    for path in missing_files:
        f.write(f'  "{path}",\n')
    f.write("];\n")

print(f"✅ Wrote {len(missing_files)} file paths to file_list.js")
```
